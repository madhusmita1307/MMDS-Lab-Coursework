```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
print(data.shape)
print(data.head())
```

```python
X = data['PetalLengthCm'].values
Y = data['PetalWidthCm'].values
```

```python
# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
```

```python
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
    m = numer / denom
    c = mean_y - (m * mean_x)
 
# Printing coefficients
print("Coefficients")
print(m, c)
```

```python
# y = mx + c 
# PetalWidthCm = (0.4164191322854012 * PetalLengthCm)-0.3665140452167275  
```

```python
# Plotting Values and Regression Line
 
max_x = np.max(X) + 0.01
min_x = np.min(X) - 0.01

x = np.linspace(min_x, max_x, 1000)
y = c + m * x
 
plt.plot(x, y, color='#58b970', label='Regression Line')
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')
 
plt.xlabel('Petal Length in cm')
plt.ylabel('Petal Width in cm')
plt.legend()
plt.show()
```

```python
# Model Evaluation
rmse = 0
for i in range(n):
    y_pred = c + m * X[i]
    rmse += (Y[i] - y_pred) ** 2
rmse = np.sqrt(rmse/n)
print("RMSE")
print(rmse)
```

```python
# Calculating R2 Score
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = c + m * X[i]
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_res/ss_tot)
print("R2 Score")
print(r2)
```
